Loan Approval Classification Dataset

Le jeu contient 14 variables et 45 000 observations

Description des Variables: 

loan_status : Résultat de la demande de prêt (1 = prêt approuvé, 0 = rejeté )

**Variables sur le demandeur :**

person_age : Âge du demandeur de prêt

person_gender : Genre du demandeur ("male", "female")

person_education : Niveau d'éducation atteint("High School", "Bachelor", "Master", "PhD", "Other")

person_income: Revenu annuel en monnaie locale

person_emp_exp : Années d'expérience professionnelle

person_home_ownership : statut logement ("RENT", "OWN", "MORTGAGE", "OTHER")

cb_person_cred_hist_length : Ancienneté de l’historique de crédit (en années)

credit_score (int) : Score de crédit du demandeur

previous_loan_defaults_on_file (cat.) : Historique de défauts de paiement	("Yes", "No")

**Variables sur le prêt :**

loan_amnt : Montant du prêt demandé

loan_intent : Motif(objet) du prêt("EDUCATION", "MEDICAL", "VENTURE", "PERSONAL", "DEBTCONSOLIDATION", "HOMEIMPROVEMENT")

loan_int_rate : Taux d’intérêt applicable (%)

loan_percent_income : Rapport (montant du prêt / revenu annuel)

Explication détaillée

**person_home_ownership**

Modalités :

"RENT" : Le demandeur loue son logement

"OWN" : Le demandeur est propriétaire de son logement

"MORTGAGE" : Le logement est en cours d’achat via un prêt hypothécaire

"OTHER" : Autre type (logement familial, logement de fonction, etc.)

**loan_intent**

"EDUCATION" : Financement des études (frais de scolarité, livres, etc.)

"MEDICAL" : Frais de santé (hospitalisation, traitement, etc.)

"VENTURE" : Projet d’entreprise ou investissement entrepreneurial

"PERSONAL" : Usage personnel (voyage, mariage, etc.)

"DEBTCONSOLIDATION" : Regroupement de dettes (consolidation de crédits existants)

"HOMEIMPROVEMENT" : Travaux dans le logement (rénovation, amélioration)

**previous_loan_defaults_on_file**

"Yes" : Le demandeur a déjà eu un prêt avec un défaut de paiement

"No" : Aucun défaut enregistré

**Utilisation des données**

Cet ensemble de données peut être utilisé à de multiples fins :

Analyse exploratoire des données (EDA) : Analyse des caractéristiques clés, des schémas de distribution et des relations pour comprendre les facteurs de risque de crédit.

Classification : Élaboration de modèles prédictifs pour classer la variable « statut du prêt » (approuvé/non approuvé) des candidats potentiels.

Régression : Élaboration de modèles de régression pour prédire la variable « score de crédit » en fonction des attributs individuels et liés au prêt.

Tenez compte des données d'origine, comme l'âge supérieur à 100 ans.

Cet ensemble de données fournit une base solide pour comprendre les facteurs de risque financier et simuler les processus de modélisation prédictive pour l'approbation des prêts et la notation de crédit.

Âge des personnes :

La distribution par âge est légèrement asymétrique à droite, la plupart des individus de l'ensemble de données ayant entre 20 et 40 ans.

Revenu des personnes :

La distribution suggère une large dispersion. La distribution est fortement asymétrique à droite, une grande partie des revenus se concentrant sur des valeurs faibles.

Expérience professionnelle des personnes :

La majorité des individus ont moins de 10 ans d'expérience

Montant du prêt :

Les montants des prêts sont concentrés sur des valeurs faibles, ce qui indique que la plupart des demandeurs sollicitent des prêts de faible montant. La distribution diminue progressivement, quelques demandeurs sollicitant des montants de prêt élevés.

Taux d'intérêt des prêts :

Les taux d'intérêt se situent généralement autour de 10 % à 15 %, ce qui correspond aux taux de prêt habituels.

Pourcentage du revenu du prêt :

Cette distribution montre que la plupart des montants de prêt représentent un faible pourcentage du revenu du demandeur, souvent inférieur à 20 %.
Quelques cas présentent un pourcentage plus élevé, suggérant des demandeurs à risque plus élevé ou à faibles revenus par rapport au montant de leur prêt.

Durée de l'historique de crédit :

La durée de l'historique de crédit culmine entre 3 et 5 ans, et moins de personnes ont un historique de crédit de plus de 10 ans.
Cette tendance peut refléter une population plus jeune ou des personnes plus novices en matière de crédit.

Score de crédit :

Les scores de crédit se situent généralement autour de la moyenne (600-700).

Niveau d’études des personnes :

La plupart des candidats sont titulaires d’un diplôme d’études secondaires, d’une licence ou d’un master, et moins nombreux sont ceux qui sont titulaires d’un diplôme d’associé ou d’un doctorat.
